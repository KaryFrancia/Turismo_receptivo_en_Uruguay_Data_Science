```python
# Se aplica get_dummies a las variables categóricas y se hace una copia del dataframe (df_turismo_dummies):
columnas_categoricas = df_turismo.select_dtypes(include=['object']).columns
df_turismo_dummies = pd.get_dummies(df_turismo, columns=columnas_categoricas, dummy_na=False)
```

```python
# Se crea una lista con el target 'GastoTotal' y con las columnas que componen al Target:
gastos = ['GastoTotal', 'GastoAlojamiento', 'GastoAlimentacion', 'GastoTransporte', 'GastoCultural', 'GastoTours', 'GastoCompras', 'GastoOtros']
# Se retira el target y los componentes del gasto total usando la lista gastos:
data = df_turismo_dummies.drop(gastos, axis=1)
data = data.apply(pd.to_numeric, errors='coerce')
# se extrae el target:
target = df_turismo_dummies["GastoTotal"]
```

```python
# se crea la siguiente función para utilizar el método de backward elimination:
def backward_elimination(data, target,significance_level = 0.05):
    features = data.columns.tolist()
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            break 
    return features
```

```python
# Las características seleccionadas son:
backward_elimination(data, target)
```

```python
# Se selecciona las columnas relevantes para el clustering
columnas_cluster = ['GastoAlojamiento', 'GastoAlimentacion', 'GastoTransporte', 'GastoCultural', 'GastoTours', 'GastoCompras', 'GastoOtros',
                  'Estadia', 'Pais', 'Motivo', 'Ocupacion', 'Alojamiento', 'TransporteLocal', 'Destino']  
df_turismo_cluster = df_turismo[columnas_cluster]

# Codificación de variables categóricas
label_encoder = LabelEncoder()

categorical_columns = ['Pais', 'Motivo', 'Ocupacion', 'Alojamiento', 'TransporteLocal', 'Destino']

for col in categorical_columns:
    df_turismo_cluster.loc[:, col] = label_encoder.fit_transform(df_turismo_cluster.loc[:, col])
```

```python
# Se estandariza las variables porque como se observó al principio algunas variables presentan mayor desviación estándar que otras
scaler = StandardScaler()
df_turismo_cluster = scaler.fit_transform(df_turismo_cluster)
```

```python
# Se inspecciona que la media y desviación estándar quedaron estandarizadas
mean_values = np.mean(df_turismo_cluster, axis=0)
std_deviation = np.std(df_turismo_cluster, axis=0)

print("Media:", mean_values)
print("Desviación Estándar:", std_deviation)
```

```python
# A través de las siguientes métricas se realiza la evaluarción para buscar el "K óptimo"
inertia_values = []
silhouette_scores = []  

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_turismo_cluster)
    inertia_values.append(kmeans.inertia_)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(df_turismo_cluster, labels)
    silhouette_scores.append(silhouette_avg)
```

```python
# Se generan los plot de silueta y se interpreta para ver con Kmeans el "K óptimo" 
X = df_turismo_cluster
labels = kmeans.labels_

k_values = [2, 3, 4, 5]

# Gráficos de silueta para diferentes valores de K
for k in k_values:
    fig, ax1 = plt.subplots(1, 1, figsize=(8, 6))

    # modelo K-Means con el valor actual de K
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    labels = kmeans.labels_

    # puntuación de silueta promedio
    silhouette_avg = silhouette_score(X, labels)

    # puntuaciones de silueta para cada punto de datos
    sample_silhouette_values = silhouette_samples(X, labels)

    y_lower = 10
    for i in range(k):
        ith_cluster_silhouette_values = sample_silhouette_values[labels == i]
        ith_cluster_silhouette_values.sort()
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        color = cm.nipy_spectral(float(i) / k)
        ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values, facecolor=color, edgecolor=color, alpha=0.7)
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10

    ax1.set_title(f"Gráfico de Silueta para K-Means con K={k}")
    ax1.set_xlabel("Puntuación de Silueta")
    ax1.set_ylabel("Etiqueta del Cluster")

    # Línea vertical para indicar la puntuación media de silueta
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    plt.show()
```
