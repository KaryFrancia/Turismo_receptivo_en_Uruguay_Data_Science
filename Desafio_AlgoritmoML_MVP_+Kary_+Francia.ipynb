## Hipótesis de interés 

1. * **Hipótesis Nula (H0)** : No hay tendencia significativa en la llegada de turistas a Uruguay en los últimos años. 
   * **Hipótesis Alternativa (H1)**: Ha habido un aumento significativo en la llegada de turistas a Uruguay en los últimos años.
2. * **H0**: El gasto promedio por visitante en Uruguay no ha cambiado con el tiempo. 
   * **H1**: El gasto promedio por visitante en Uruguay ha experimentado cambios significativos.
3. * **H0**: Las nacionalidades de los turistas están igualmente distribuidas en Uruguay. 
   * **H1**: Algunas nacionalidades son significativamente más frecuentes que otras entre los turistas en Uruguay.
4. * **H0**: No hay diferencia significativa en la elección de destinos turísticos en Uruguay. 
   * **H1**: Existen diferencias significativas en la elección de destinos turísticos en Uruguay.
5. * **H0**: La duración promedio de la estadía de los turistas en Uruguay es constante y no ha cambiado significativamente en los últimos años.
   * **H1**: La duración promedio de la estadía de los turistas en Uruguay ha experimentado cambios significativos en los últimos años.
6. * **H0**: Todos los motivos de viaje son igualmente frecuentes entre los turistas en Uruguay.
   * **H1**: Algunos motivos de viaje son más comunes que otros entre los turistas en Uruguay.
7. * **H0**: Los componentes del gasto de los turistas en Uruguay están igualmente distribuidos y no hay diferencias significativas en la contribución de cada componente al gasto total.
   * **H1**: Existen diferencias significativas en la contribución de los componentes del gasto de los turistas en Uruguay, y algunos componentes son más prominentes que otros en términos de su contribución al gasto total.
8. * **H0**: No hay relación significativa entre el motivo principal del viaje y la duración promedio de la estadía.
    * **H1**: Existe una relación significativa entre el motivo principal del viaje y la duración promedio de la estadía.
9. * **H0**: No hay relación significativa entre la duración promedio de la estadía y el gasto promedio.
    * **H1**: Existe una relación significativa entre la duración promedio de la estadía y el gasto promedio.
10. * **H0**: La distribución del gasto es la misma en todas las categorías.
    * **H1**: La distribución del gasto varía según el motivo del viaje, con ciertas categorías que son más prominentes para ciertos grupos.
11. * **H0**: Los turistas que viajan por segunda residencia no gastan más en promedio que los turistas que viajan por placer.
    * **H1**: Los turistas que viajan por segunda residencia gastan más en promedio que los turistas que viajan por placer.
12. * **H0**: No se pueden identificar grupos de turistas con patrones de gasto similares en función de su motivo de viaje.

# MODELO RandomForestRegressor

<hr/>
<div class="alert alert-success alertsuccess" style="margin-top: 20px">
    <p>  Se utiliza el  modelo de regresión RandomForestRegressor para intentar predecir el "Gasto Total" realizado por los turistas. </p>

<hr/>


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Se separa la variable a predecir "y" del resto de variables
X = df_turismo.drop('GastoTotal', axis=1)
y = df_turismo['GastoTotal']
```

```python
# Se aplica One-Hot Encoding a las variables categóricas
X_encoded = pd.get_dummies(X, columns=['Lugar Ingreso', 'Transporte Internacional de Ingreso', 'Pais', 'Residencia', 'Motivo', 'Ocupacion', 'Estudio',
       'Localidad', 'Departamento', 'Otro Departamento', 'Otra Localidad',
       'Alojamiento', 'TransporteLocal', 'Lugar Egreso',
       'Transporte Internacional de Egreso', 'Destino'])

# Se divide el dataset en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Se extrae el mes y el día como características numéricas (características adicionales de las columnas de fecha)
X_train['mes_ingreso'] = X_train['FechaIngreso'].dt.month
X_train['dia_ingreso'] = X_train['FechaIngreso'].dt.day
X_train['mes_egreso'] = X_train['FechaEgreso'].dt.month
X_train['dia_egreso'] = X_train['FechaEgreso'].dt.day

X_test['mes_ingreso'] = X_test['FechaIngreso'].dt.month
X_test['dia_ingreso'] = X_test['FechaIngreso'].dt.day
X_test['mes_egreso'] = X_test['FechaEgreso'].dt.month
X_test['dia_egreso'] = X_test['FechaEgreso'].dt.day
```

```python
# Se escala con MinMax para normalizar las variables
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train.drop(columns=['FechaIngreso', 'FechaEgreso']))
X_test_scaled = scaler.transform(X_test.drop(columns=['FechaIngreso', 'FechaEgreso']))
```

```python
# Se crea y entrena el modelo de Random Forest Regressor
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train_scaled, y_train)

# Se realizan las predicciones en el conjunto de prueba
y_pred = rf_model.predict(X_test_scaled)

# Evaluación del modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Error Cuadrático Medio: {mse}")
print(f"R-cuadrado: {r2}")
```

Es posible que haya un sobreajuste (overfitting) en el modelo, debido a que se obtuvo un R2 muy alto en el conjunto de prueba. Se puede inferir que el modelo se ajusta demasiado bien a los datos de entrenamiento y podría estar capturando el ruido en lugar de los patrones subyacentes.</p>

<hr/>
<div class="alert alert-success alertsuccess" style="margin-top: 20px">
    <p> Al parecer el modelo podría estar presentando un overfitting. Por ello, se realiza una Validación Cruzada y el Ajuste de Hiperparámetros, a fin de encontrar un mejor modelo. Para tal efecto, no se utiliza  GridSearchCV, ya que si bien este realiza una búsqueda más exhaustiva puede ser más costoso computacionalmente, especialmente si hay muchos hiperparámetros y valores para explorar. Por lo tanto, en este caso se utiliza RandomizedSearchCV, el cual realiza una búsqueda más eficiente al seleccionar aleatoriamente un subconjunto de combinaciones de hiperparámetros para la evaluación.  </p>
</div>
<hr/>

```python
# Se define la siguiente cuadrícula de hiperparámetros:

param_grid = {
    'n_estimators': [int(x) for x in np.linspace(start=200, stop=2000, num=10)],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [int(x) for x in np.linspace(10, 110, num=11)],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Se utiliza luego RandomizedSearchCV , con n_jobs igual a 1 debido al rendimiento del equipo. 
random_search = RandomizedSearchCV(estimator=rf_model, param_distributions=param_grid, n_iter=20, cv=5, verbose=2, random_state=42, n_jobs=1)

# Se entrena el modelo. 
random_search.fit(X_train_scaled, y_train)
```
